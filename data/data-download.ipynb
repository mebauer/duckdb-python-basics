```python
# download CSV from source as a CSV
url = "https://data.cityofnewyork.us/api/views/bs59-f3nu/rows.csv"

output_path = "green_infrastructure.csv"

response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    with open(output_path, 'wb') as f:
        f.write(response.content)
    print("CSV file downloaded successfully.")
else:
    print("Failed to download CSV file. Status code:", response.status_code)    
```

# Download Data Dictionary

```python
url = "https://data.cityofnewyork.us/api/views/spjh-pz7h/files/58a6d0ec-f6a0-4ad4-b410-9aa80140d8a4\
?download=true&filename=DEPGreenInfrastructure_Data_Dictionary.xlsx"

output_path = "data-dictionary.xlsx"

response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    with open(output_path, 'wb') as f:
        f.write(response.content)
    print("Excel file downloaded successfully.")
else:
    print("Failed to download Excel file. Status code:", response.status_code)    
```
